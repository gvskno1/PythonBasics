# 4. Statistical Functions 

```python
# Find mean
s = np.array([1, 2, 3, 4])
mean = np.mean(s)
print("Mean of s:", mean) # find mean , average
```

```python
# Finding Variants ----- (x1-mean)square+..../totale 
# Variance i sthe measure of how far a set of numbers are spread out from their average value
s = np.array([1,2,3,4]) 
vars = np.var(s)
print("Variants;", vars)
```

```python
# Finding standard deviation  ----- (x1-mean)square+..../totale
# Standard deviation is a measure of the amount of variation or dispersion of a set of values.
s = np.array([1,2,3,4]) 
std = np.std(s)
print("Standard deviation;", std)
```

```python
# Finding Minimum element of an array 
Minimum = np.min(s)
print("Min of s:", Minimum)
```

```python
# Maximum element of an array 
maximum = np.max(s)
print("Max of s:", maximum)
```

# 7. Boolean and logical functions

```python
# check if all elements are true 
logical_test = np.array([True, False, True])
all_true = np.all(logical_test) # Check if all are True 
print(" All elements True:", all_true )
```

```python
# Check if any elements are true 
logical_test = np.array([True, False, True])
any_true = np.any(logical_test) # Check if all are True 
print(" Any elements True:", any_true )
```

# 8. Set Operations 

```python
# Intersection of two arrays 
set_a = np.array([1, 2, 3, 4])
set_b = np.array([3, 4, 5, 6])
intersection = np.intersect1d(set_a, set_b)
print("Intersection of a and b:", intersection)
```

```python
# Union of two arrays 
union = np.union1d(set_a, set_b)
print("union of a nad b:", union)
```

# 9. Array Attribute Functions 

```python
# Array Attribute 
a = np.array([1,2,3])
shape = a.shape # Shape of the array 
size = a.size # Number of elements 
dimensions = a.ndim # Number of dimensions 
dtype = a.dtype # Data type of elements

print("Shape of a:", shape)
print("Size of a:", size)
print("Dimensions of a:", dimensions)
print("Data type of a:", dtype)

```
