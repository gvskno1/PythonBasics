# Type Casting

```python
int(2.3)    # float to int 
```

```python
int(True) # bool to int 
```

```python
int(False) # bool to int
```

```python
int(1+2j) # complex to int  -- will raise error
```

```python
int('10')
```

```python
int('ten')
```

```python
int(2.3, 4.5)   # float to int  -- will raise error
```

```python
float(10) # int to float
```

```python
float(True) # bool to float
```

```python
float(1+2j) # complex to float  -- will raise error
```

```python
float('10')
```

```python
float('ten')
```

# Other datatype to complex

```python
complex(10) # int to complex
```

```python
complex(10,20) # int to complex with real and imaginary parts
```

```python
complex(9.7) # float to complex
```

```python
complex(9.7, 8.5) # float to complex with real and imaginary parts
```

```python
complex(1, 2.3)
```

```python
complex(False, True)
```

```python
complex('10')
```

```python
complex('10', '20')
```

```python
complex(10,'20') # int and string to complex
```

```python
complex('10', 20) # string and int to complex
```

```python
bool(1) # int to bool
```

```python
bool(0) # int to bool
```

```python
bool( )  # empty value to bool
```

```python
bool(2.3) # float to bool
```

```python
bool (2.3, 3.4) # float to bool  -- will raise error
```

```python
bool('ten') # string to bool
```

```python
bool(1+2j) # complex to bool
```
