# Intersection &

```python
s5 = {1,2,3,4,5}
s6 = {4,5,6,7,8}
s7 = {8,9,10}
```

```python
s5.intersection(s6)# prints only the common elements
```

```python
s6 & s7 
```

```python
s5.intersection(s7) # No Common elements so will give empty set
```

# Symmetric difference 

```python
s5 = {1,2,3,4,5}
s6 = {4,5,6,7,8}
s7 = {8,9,10}
```

```python
s5.symmetric_difference(s6)# the common items will not be printed and rest of the uncommon will be printed from both the sets. 

print(s5)
print(s6)
print(s7)
```
